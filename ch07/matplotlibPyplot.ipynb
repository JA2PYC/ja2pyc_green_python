```python
# Matplotlib Plotlib Use xlsx Data with Pandas
import pandas as pd
import matplotlib.pyplot as plt
a = pd.read_excel('data1.xlsx')
plt.plot(a['kor'])
plt.show()
```

```python
# Matplotlib Plotlib plot Line Style, Marker
import pandas as pd
import matplotlib.pyplot as plt
a = pd.read_excel('data1.xlsx')
plt.plot(a['kor'], 'k--', marker='o')
plt.show()
```

```python
# Matplotlib Pyplot plot Label
import matplotlib.pyplot as plt
import pandas as pd
a = pd.read_excel('data1.xlsx')
plt.plot(a['kor'], 'k--', marker='.', label='Korean')
plt.plot(a['math'], 'r-.', marker='^', label='Math')
plt.legend()
plt.show()
```

```python
# Matplotlib Pyplot scatter
import matplotlib.pyplot as plt
a = [6,2,1,1,4,4,6,4,7,7]
b = [5,4,2,9,9,0,10,3,1,2]
plt.scatter(a,b)
plt.show()
```

```python
# Matplotlib with numpy
import numpy as np
import matplotlib.pyplot as plt
a = np.arange(0, 10, 1)
b = np.random.randint(0, 10, size=10)
x = np.linspace(1, 10, 10)
print (a)
print (b)
print (x)
plt.scatter(x,a)
plt.scatter(x,b)
plt.show()
```

```python
# Matplotlib Pyplot method bar
import matplotlib.pyplot as plt
x = ['A', 'B', 'C']
y = [100, 150, 120]
plt.bar(x,y)
plt.show()
```

```python
# Matplotlib Pyplot method barh
import matplotlib.pyplot as plt
x = ["D", "E", "F"]
y = [100, 150, 130]
plt.barh(x, y, color='#a0a0a0')
plt.show()
```

```python
# Matplotlib Pyplot method hist
import matplotlib.pyplot as plt
x = [1, 1, 1, 2, 3, 2, 4, 5, 5, 5, 5, 5]
plt.hist(x, bins=5, color='green', align='mid')
plt.show()
```

```python
# Matplotlib Pyplot method title xlabel, ylabel
import matplotlib.pyplot as plt
x = ['G', 'H', 'I']
y = [95, 80, 92]
plt.bar(x,y, color = 'blue')
plt.title('Group Score')
plt.xlabel('Group')
plt.ylabel('Score')
plt.show()
```

```python
# Matplotlib Pyplot method subplots
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
fig, axs = plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,1].scatter(x,y)
axs[1,0].bar(x,y)
axs[1,1].barh(x,y)
fig.suptitle('Drawing Graph')
plt.show()
```

```python
# Matplotlib Set Font
plt.rc('font', family='Noto Sans KR')
x = [1, 2, 3, 4]
y = [10, 4, 15, 9]
plt.plot(x, y, label = '')
plt.title(' ')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.show()
```
