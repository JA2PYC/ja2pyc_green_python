```python
# Question 3
import matplotlib.pyplot as plt
import seaborn as sns
species_data = sns.load_dataset('iris')
sns.barplot(y='petal_length', hue='species', data= species_data)
plt.title('Iris Graph')
plt.show()
species_data
```

```python
# Question 5
import matplotlib.pyplot as plt
import seaborn as sns
penguins_data = sns.load_dataset('penguins')
sns.regplot(x= 'flipper_length_mm', y = 'body_mass_g', data = penguins_data)
plt.title('Penguins Graph')
plt.show()          
penguins_data
```
