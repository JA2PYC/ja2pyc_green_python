```python
# Seaborn Method scatterplot
import matplotlib.pyplot as plt
import seaborn as sns
tips_data = sns.load_dataset('tips')
sns.scatterplot(x = 'total_bill', y='tip', hue='day', data=tips_data)
plt.title('Total Bill & Tips')
plt.show()
```

```python
# Seaborn Method counterplot
import matplotlib.pyplot as plt
import seaborn as sns
tips_data = sns.load_dataset('tips')
# sns.counterplot(x='day', data = tips_data)
sns.countplot(x='day', data = tips_data)
plt.title('Day Counter')
plt.show()
tips_data
```

```python
# Seaborn Method counterplot with hue
import matplotlib.pyplot as plt
import seaborn as sns
tips_data = sns.load_dataset('tips')
sns.countplot(x='sex', hue='day', data = tips_data)
plt.title('Counter Data')
plt.show()
tips_data
```

```python
# Seaborn Method barplot
import matplotlib.pyplot as plt
import seaborn as sns
tips_data = sns.load_dataset('tips')
sns.barplot(x='sex', y='tip', hue='day', data=tips_data)
plt.title('Bar Data')
plt.xlabel('Gender')
plt.ylabel('Tip')
plt.show()
```

```python
# Seaborn Method hisplot
import matplotlib.pyplot as plt
import seaborn as sns
tips_data = sns.load_dataset('tips')
sns.histplot(x='tip', kde=True, data=tips_data)
plt.title('Hist Data')
plt.xlabel('Tip')
plt.ylabel('Value')
plt.show()
```

```python
# Seaborn Method pairplot
import matplotlib.pyplot as plt
import seaborn as sns
tips_data = sns.load_dataset('tips')
sns.pairplot(hue='day', data=tips_data)
plt.show()
```

```python
# Seaborn Method regplot
import matplotlib.pyplot as plt
import seaborn as sns
tips_data = sns.load_dataset('tips')
sns.regplot(x='total_bill', y='tip', data=tips_data)
plt.title('regplot Graph')
plt.show()
```

```python
# Seaborn Method heatmap
import matplotlib.pyplot as plt
import seaborn as sns
tips_data = sns.load_dataset('tips')
df = tips_data[['total_bill','tip','size']].corr()
sns.heatmap(annot = True, fmt='.2f', cbar = True, data=df)
plt.show()
```

```python
# Seaborn Penguins barplot
import matplotlib.pyplot as plt
import seaborn as sns
penguins_data = sns.load_dataset('penguins')
sns.barplot(x='species', y='body_mass_g', hue='sex', data=penguins_data)
plt.title('Penguins Graph')
plt.show()
```

```python
# Seaborn Penguins heatmap
import matplotlib.pyplot as plt
import seaborn as sns
penguins_data = sns.load_dataset('penguins')
df = penguins_data[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].corr()
sns.heatmap(annot=True, data = df)
plt.title('Penguins Heatmap')
plt.show()
penguins_data
```

```python
# Seaborn Penguins regplot
import matplotlib.pyplot as plt
import seaborn as sns
penguins_data = sns.load_dataset('penguins')
sns.regplot(x='body_mass_g', y='flipper_length_mm', data =penguins_data)
plt.title('Penguins Regplot')
plt.show()
penguins_data
```
