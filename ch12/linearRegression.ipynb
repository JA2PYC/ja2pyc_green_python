```python
# Linear Regression
# Import Data Library
import numpy as np

# Import Graph Library
import matplotlib.pyplot as plt

# Import Linear Regression Model
import sklearn.linear_model as skl
```

```python
# Set Data X [[]], Y []
data_x = np.array([[2.1], [10], [3], [1], [3.5], [5], [8]], dtype=float)
data_y = np.array( [64.9, 292.6, 85.9, 30.92, 110.5, 163.4, 230.1], dtype=float)
print (type(data_x), data_x)
print (type(data_y), data_y)
```

```python
# Data Graph
plt.scatter(data_x, data_y)
plt.show()
```

```python
# Linear Regression Model
skl_model = skl.LinearRegression()
skl_model = skl_model.fit(data_x, data_y)
skl_model
```

```python
# Predict by Linear Model
data_predict = [[7], [12]]
skl_result = skl_model.predict(data_predict)
print(skl_result)
```

```python
# Compare Sample Data and Predict Data
plt.scatter(data_x, data_y, label='Training Data')
plt.scatter(data_predict, skl_result, marker='v', label='Predict Data')
plt.legend()
plt.show()
```

```python
# Linear Model coef intercept
line_x = np.array([1,12])
line_y = line_x*skl_model.coef_ + skl_model.intercept_
plt.plot(line_x, line_y)
plt.scatter(data_x, data_y, label='Training Data')
plt.scatter(data_predict, skl_result, marker='v', label='Predict Data')
plt.show()
```
