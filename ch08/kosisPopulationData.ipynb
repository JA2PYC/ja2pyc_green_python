```python
# Population and Birth Data use Subplot
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('./data/총인구수출생아수.xlsx')
plt.rc('font', family='Noto Sans Kr')
fig, axs = plt.subplots()
axs.set_xlabel('연도')
axs.set_ylabel('총인구수')
axs.plot(df['연도'], df['총인구수']/10000)

axs_sub = axs.twinx()
axs_sub.set_ylabel('출생아수', color='red')
axs_sub.plot(df['연도'], df['출생아수']/10000, color='red')
axs_sub.tick_params(axis='y', labelcolor='red')
plt.title('총인구수 출생아수 Subplot')
plt.show()
df.head(5)
```
