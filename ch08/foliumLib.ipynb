```python
# Park Data ulsan
import pandas as pd
import folium as fl
df = pd.read_excel('./data/전국도시공원정보표준데이터_신버전.xlsx')
df = df.dropna(subset=['위도', '경도', '소재지지번주소'])
df = df.query("소재지지번주소.str.contains('울산')", engine='python')
mymap = fl.Map(location=[35.541970, 129.338271], zoom_start=11)
fl.Marker([35.541970, 129.338271], popup='학원').add_to(mymap)
for i in range(len(df)):
    fl.CircleMarker([float(df.iloc[i,5]), float(df.iloc[i,6])], popup=df.iloc[i,1]).add_to(mymap)
mymap
```

```python
# Population Chart
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('./data/주민등록인구.xlsx', thousands=',')
userInput = input('동이름을 입력하세요>')
df = df.query('행정기관.str.contains("'+ userInput + '")', engine='python')
df = df.iloc[0:1, 4:]
arr = np.array(df, dtype=int).flatten()
df_x = np.array(df.columns)

plt.rc('font', family='Noto Sans Kr')
plt.figure(figsize=(10,7))
plt.title(userInput + ' 인구 그래프')
plt.plot(df_x, arr, marker='*')
plt.show()
```

```python
# Pandas use T
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('./data/주민등록인구.xlsx', thousands=',')
userInput = input('동명을 입력해주세요>')
df = df.query('행정기관.str.contains("' + userInput + '")', engine = 'python')
df = df.iloc[0:1, 4:]
df = df.T
df = df.astype(int)
plt.rc('font', family='Noto Sans Kr')
plt.figure(figsize = (10,7))
plt.title(userInput + '의 인구 데이터')
plt.plot(df)
plt.show()
```

```python
# Population Multiple Data
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('./data/주민등록인구.xlsx', thousands=',')
userInput = input('동이름을 입력해주세요>')
df = df.query('행정기관.str.contains("' + userInput + '")', engine='python')
df_column = df.iloc[0:, 4:].columns
plt.rc('font', family='Noto Sans Kr')
plt.figure(figsize = (10, 7))
plt.title(userInput + '의 인구 데이터')
for i in range(len(df)):
    df_i = df.iloc[i:i+1, 4:]
    df_i = np.array(df_i, dtype =int).flatten()
    plt.plot(df_column, df_i, label=df.iloc[i, 1])
plt.legend()
plt.show()
```
