```python
# Pandas Operator
import pandas as pd
a = pd.read_excel('data.xlsx')
a['국어'] >= 80
```

```python
# Operator Query []
import pandas as pd
a = pd.read_excel('data.xlsx')
a[a['국어'] >= 80]
```

```python
# Operator Query query('')
import pandas as pd
a = pd.read_excel('data.xlsx')
b = a.query('국어 >= 80')
b
```

```python
# Operator Query [] & []
import pandas as pd
a = pd.read_excel('data.xlsx')
b = a[(a['국어'] >= 80) & (a['수학'] >= 90)]
b
```

```python
# Operator Query [] | []
import pandas as pd
a = pd.read_excel('data.xlsx')
b = a[(a['국어'] >= 80) | (a['수학'] >= 90)]
b
```

```python
# Operator sort_values()
import pandas as pd
a = pd.read_excel('data.xlsx')
b = a.sort_values(by=['이름'])
b
```

```python
# Operator sort_values(by, inplace)
import pandas as pd
a = pd.read_excel('data.xlsx')
a.sort_values(by=['이름'], inplace=True)
a
```

```python
# Operator sort_values(by, inplace) reset_index(drop, inplace)
import pandas as pd
a = pd.read_excel('data.xlsx')
a.sort_values(by=['이름'], inplace=True)
a.reset_index(drop=True, inplace = True)
a
```
