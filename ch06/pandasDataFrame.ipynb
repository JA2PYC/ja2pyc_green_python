```python
# Pandas Data Frame
import pandas as pd
a = pd.DataFrame([['Kim', 27, 92], ['Lee', 33, 98], ['Park', 19, 87]])
print (type(a), a)
a
```

```python
# Dictionary Data Frame
import pandas as pd
a = pd.DataFrame({'name' : ['Kim', 'Lee', 'Park'],
                  'age' : [27, 33, 19],
                  'score' : [92, 98, 87]})
print (type(a), a)
a
```

```python
# CSV Data Frame
import pandas as pd
a = pd.read_csv('data.csv', encoding='UTF-8')
print (type(a), a)
a
```
