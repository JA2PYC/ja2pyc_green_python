```python
# Import Data Library
import numpy as np
import pandas as pd

# Import Machine Learning Library
import sklearn.tree as skt
import sklearn.neighbors as skn
import sklearn.ensemble as ske

# Import Accuracy Evaluation Library
import sklearn.metrics as skm
```

```python
# Load Data Frame
model_df = pd.read_excel('./data/사과배복숭아-학습.xlsx')
print (model_df.isna().sum())
filled_model_data = model_df.fillna(0)
print (filled_model_data.isna().sum())
convert_map = {'사과' : 0, '배' : 1, '복숭아' : 2}
converted_model_data = filled_model_data.copy()
converted_model_data['fruit'] = converted_model_data['fruit'].map(convert_map)
converted_model_data
```

```python
# Set Model Data
model_data = converted_model_data.iloc[:, 1:]
model_answer = converted_model_data.iloc[:, 0]
```

```python
# Tree Model
tree_model = skt.DecisionTreeClassifier(random_state=0)
tree_model = tree_model.fit(model_data, model_answer)
tree_model
```

```python
# K Nnearest Neighbors Model
knn_model = skn.KNeighborsClassifier(n_neighbors=5)
knn_model = knn_model.fit(model_data, model_answer)
knn_model
```

```python
# Random Forest Model
rf_model = ske.RandomForestClassifier(n_estimators=7)
rf_model = rf_model.fit(model_data, model_answer)
rf_model
```

```python
# Predict Data
predict_df = pd.read_excel('./data/사과배복숭아-테스트.xlsx')
filled_predict_data = predict_df.fillna(0)
converted_predict_data = filled_predict_data.copy()
converted_predict_data['fruit'] = converted_predict_data['fruit'].map(convert_map)
predict_data = converted_predict_data.iloc[:, 1:]
predict_answer = converted_predict_data.iloc[:, 0]
```

```python
# Predict Result by Tree Model
tree_predict_result = tree_model.predict(predict_data)
tree_accuracy_score = skm.accuracy_score(predict_answer, tree_predict_result)
tree_recall_score = skm.recall_score(predict_answer, tree_predict_result, average=None)
tree_precision_score = skm.precision_score(predict_answer, tree_predict_result, average=None)
print ('정확도 :', tree_accuracy_score)
print ('재현율 :', tree_recall_score)
print ('정밀도 :', tree_precision_score)
```

```python
# Predict Result by K Nearest Neighbors Model
knn_predict_result = knn_model.predict(predict_data)
knn_accuracy_score = skm.accuracy_score(predict_answer, knn_predict_result)
knn_recall_score = skm.recall_score(predict_answer, knn_predict_result, average = None)
knn_precision_score = skm.precision_score(predict_answer, knn_predict_result, average = None)
print ('정확도 :', knn_accuracy_score)
print ('재현율 :', knn_recall_score)
print ('정밀도 :', knn_precision_score)
```

```python
# Predict Result by Random Forest Model
rf_predict_result = rf_model.predict(predict_data)
rf_accuracy_score = skm.accuracy_score(predict_answer, rf_predict_result)
rf_recall_score = skm.recall_score(predict_answer, rf_predict_result, average=None)
rf_precision_score = skm.precision_score(predict_answer, rf_predict_result, average=None)
print ('정확도 :', rf_accuracy_score)
print ('재현율 :', rf_recall_score)
print ('정밀도 :', rf_precision_score)
```
