```python
# Import Data Library
import numpy as py
import pandas as pd

# Import Graph Library
import matplotlib.pyplot as plt
import seaborn as sb

# Import Machine Learning Library
from sklearn.ensemble import RandomForestClassifier as rfc

# Import Accuracy Score Library
from sklearn.metrics import accuracy_score as asc
```

```python
# Load Data
df = pd.read_excel('./data/사과배복숭아-학습.xlsx')
filled_data = df.fillna(0)
convert_map = {'사과' : 0, '배': 1, '복숭아':2}
converted_data = filled_data.copy()
converted_data['fruit'] = converted_data['fruit'].map(convert_map)
converted_data
```

```python
# Set Model Data
model_data = converted_data.iloc[:, 1:]
model_answer = converted_data.iloc[:, 0]
```

```python
# Random Forest Model
rfc_model = rfc(n_estimators=3)
rfc_model = rfc_model.fit(model_data, model_answer)
rfc_model
```

```python
# Set Predict Data
predict_df = pd.read_excel('./data/사과배복숭아-테스트.xlsx')
filled_predict_data = predict_df.fillna(0)
converted_predict_data = filled_predict_data.copy()
converted_predict_data['fruit'] = filled_predict_data['fruit'].map(convert_map)
predict_data = converted_predict_data.iloc[:, 1:]
predict_answer = converted_predict_data.iloc[:, 0]
predict_data
```

```python
# Predict by Random Forest Model
predict_result = rfc_model.predict(predict_data)
predict_result
```

```python
# Accuracy Score
rfc_score = asc(predict_answer, predict_result)
print ('RFC Score :', rfc_score)
for i in range(len(predict_result)):
    i_result = False
    if predict_result[i] == predict_answer[i] :
        i_result = True
    print ('예측 :', predict_result[i], '정답 :', predict_answer[i], '예측결과 :', i_result)
```
